```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
import plotly.express as px

%matplotlib inline 
```

```python
credits_df = pd.read_csv("C:/Users/Saruna/Desktop/Task files/movies/credits.csv")
credits_df.head(10)
```

```python
titles_df = pd.read_csv("C:/Users/Saruna/Desktop/Task files/movies/titles.csv")
titles_df.head(10)
```

```python
ratings_df = pd.read_csv("C:/Users/Saruna/Desktop/Task files/movies/ratings.csv")
ratings_df.head(10)
```

```python
# common columns between titles and ratings
common_columns = list(set(titles_df.columns).intersection(set(ratings_df.columns)))
print("Common columns:", common_columns)

# Select necessary columns from titles and ratings
titles_selected = titles_df[['id', 'title', 'type', 'description', 'release_year', 'age_certification', 'runtime', 'imdb_id']]
ratings_selected = ratings_df[['id', 'imdb_score', 'imdb_votes']]

# Merge titles and ratings on the 'id' column including only common columns
merged_df = pd.merge(titles_selected, ratings_selected, on='id')

# Drop duplicates based on 'id'
deduplicated_df = merged_df.drop_duplicates(subset=['id'])


print("Deduplicated DataFrame:")
deduplicated_df.head(10)
```

```python
# Longest and shortest runtimes among movies
longest_runtime = deduplicated_df['runtime'].max()
shortest_runtime = deduplicated_df[deduplicated_df['runtime'] > 0]['runtime'].min()

print(f"Longest Runtime: {longest_runtime} minutes")
print(f"Shortest Runtime: {shortest_runtime} minutes")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average runtime per release year for movies and shows
avg_runtime_per_year = deduplicated_df.groupby(['release_year', 'type'])['runtime'].mean().reset_index()

# Plot the results
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_runtime_per_year, x='release_year', y='runtime', hue='type', marker='*')

plt.title('Average Runtime per Release Year for Movies and Shows')
plt.xlabel('Release Year')
plt.ylabel('Average Runtime (minutes)')
plt.legend(title='Type')
plt.show()
```

```python
# Calculate the length of each movie title and add as a new column
deduplicated_df['title_length'] = deduplicated_df['title'].apply(len)

# Sort DataFrame by title_length to find the longest and shortest titles
df_sorted = deduplicated_df.sort_values(by='title_length', ascending=False)

# Extract only the title and title_length columns for the longest and shortest titles
longest_titles = df_sorted.head(10)[['title', 'title_length']]
shortest_titles = df_sorted.tail(10)[['title', 'title_length']]

# Display 10 longest and 10 shortest movie titles along with their lengths
print("Longest Movie Titles:")
print(longest_titles)

print("\nShortest Movie Titles:")
print(shortest_titles)
```

```python
# Count the number of movies per genre
titles_df['genres'] = titles_df['genres'].str.split(', ')
genres_exploded = titles_df.explode('genres')
genres_count = genres_exploded['genres'].value_counts().reset_index()
genres_count.columns = ['genre', 'num_movies']

# Plot number of movies per genre
plt.figure(figsize=(14, 8))
sns.barplot(data=genres_count, x='genre', y='num_movies')
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
# Combine deduplicated list with credits table
combined_df = pd.merge(deduplicated_df, credits_df, on='id')

# Find the actor that starred in the most movies by release year
actor_movies = combined_df.groupby(['release_year', 'name']).size().reset_index(name='num_movies')
top_actors = actor_movies.loc[actor_movies.groupby('release_year')['num_movies'].idxmax()]

# Plot the number of movies the top actor starred in each year
plt.figure(figsize=(14, 8))
sns.barplot(data=top_actors, x='release_year', y='num_movies', hue='name')
plt.title('Number of Movies the Top Actor Starred in by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Actor')
plt.show()
```
